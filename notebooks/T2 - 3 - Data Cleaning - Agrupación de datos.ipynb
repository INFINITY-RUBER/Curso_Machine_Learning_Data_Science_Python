# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
from string import ascii_letters 
```

```python

#print (ascii_letters)
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,500):
    gender_data.append(np.random.choice(gender)) # añade un elemento aleatorio
    income_data.append(np.random.choice(income))
```

```python
print("gender_data:",gender_data[0:10])
print("income_data:",income_data[0:10])
```

```python
# Z -> N(0,1)
# N(m, s) --> m + s * Z
height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.rand(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income
    }
)
```

```python
data.shape
```

# Conjunto de entrenamiento y conjunto de testing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Dividir utilizando la distribución normal

```python
a = np.random.randn(len(data)) # .randn es la distrubucion normal
```

```python
plt.hist(a)
```

```python
check = (a < 0.75) # CHEQUEA en  un array de True o False
```

```python
print("check.shape:",check.shape)
print("check:",type(check))
```

```python
conta_true = 0
conta_False = 0
chec = []
for i in check:
    if i == True:
        conta_true += 1
        chec.append(1)
    else:
        conta_False += 1
        chec.append(0)
        
print("True:",conta_true, "False:",conta_False)   
chec
```

```python
plt.hist(chec)
```

```python
training = data[check]
testing = data[~check]
```

```python
len(training)
```

```python
len(testing)
```
