# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../datasets/movies/movies.csv", sep=";")
data
```

```python
# QUITAR LA PRIMERA COLUMNA ID
movies = data.columns.values.tolist()[1:]# combierto en lista y quita el primero
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1)# para la distancia de manthatam

dd2 = distance_matrix(data[movies], data[movies], p=2) # para la distacia de Euclidia

dd10 = distance_matrix(data[movies], data[movies], p=10)# otra distancia cualquiera
```

```python
print(dd1)
type(dd2) 
```

```python
# funcion para convertir los array a DataFrame
def dm_to_df(dd, col_name):
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
# el nombre de las row van a tener los indexes de las columnas ,  y las index de columnans se llamaran = indexes de las columnas
```

```python
dm_to_df(dd1, data["user_id"])# nombre de la culumna de los usuarios
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

```python
df[11]=df[1]+df[10] # CREA UNA COLUMNA [11] y suma columna [1]+ columna[10]
df.loc[11]=df.loc[1]+df.loc[10] # CREA UNA FILA [11] y suma FILA [1]+ FILA [10]

# ([1,10,0.7,2])>>>  id1, id2, d, n_elementos_en_cluster -> 11.
Z.append([1,10,0.7,2])
df
```

### REORDENACION DE MATRIZ 
df.loc [FILAS][COLUMNAS]

```python
#   df.loc [FILAS][COLUMNAS]

for i in df.columns.values.tolist():
    df.loc[11][i] = min(df.loc[1][i], df.loc[10][i])# FILA[11] sera el minimo entre toda la FILA 1 vS toda la FILA 10
    df.loc[i][11] = min(df.loc[i][1], df.loc[i][10])# COLUMN[11] sera el minimo entre toda la COLUMN 1 vS toda la COLUMN 10
df # reace la matris de distancia
```

# Clustering jerrquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], "average")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "complete")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], method="single", metric="cosine")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```
